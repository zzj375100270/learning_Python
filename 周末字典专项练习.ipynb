# ç¬¬ä¸‰éƒ¨åˆ†ï¼šç»¼åˆå°é¡¹ç›®

## é¡¹ç›®1ï¼šç®€æ˜“å­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿ

**éœ€æ±‚**ï¼š
1. å¯ä»¥æ·»åŠ å­¦ç”Ÿå’Œæˆç»©
2. å¯ä»¥æŸ¥è¯¢å­¦ç”Ÿæˆç»©
3. å¯ä»¥è®¡ç®—å¹³å‡åˆ†
4. å¯ä»¥æ‰¾å‡ºæœ€é«˜åˆ†å’Œæœ€ä½åˆ†

**æ•°æ®ç»“æ„è®¾è®¡**ï¼š
ä½¿ç”¨å­—å…¸å­˜å‚¨ï¼Œé”®ä¸ºå­¦ç”Ÿå§“åï¼Œå€¼ä¸ºæˆç»©

**å®ç°æ­¥éª¤**ï¼š
1. å…ˆå®šä¹‰éœ€è¦çš„å‡½æ•°
2. ç¼–å†™æ¯ä¸ªå‡½æ•°çš„å…·ä½“å®ç°
3. æµ‹è¯•å‡½æ•°çš„æ­£ç¡®æ€§


````python
# é¡¹ç›®1ï¼šå­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿ - åˆå§‹åŒ–

# åˆå§‹åŒ–æˆç»©å­—å…¸
grades = {}

# åŠŸèƒ½1ï¼šæ·»åŠ å­¦ç”Ÿæˆç»©
def add_student(name, score):
    """æ·»åŠ æˆ–æ›´æ–°å­¦ç”Ÿæˆç»©"""
    grades.update(name,score)
    print(f"å·²æ·»åŠ /æ›´æ–°: {name} -> {score}")

# åŠŸèƒ½2ï¼šæŸ¥è¯¢æˆç»©
def query_score(name):
    """æŸ¥è¯¢å­¦ç”Ÿæˆç»©"""
    result = grades.get(name,"æŸ¥æ— æ­¤äººï¼")
    return result

# åŠŸèƒ½3ï¼šè®¡ç®—å¹³å‡åˆ†
def calculate_average():
    """è®¡ç®—æ‰€æœ‰å­¦ç”Ÿçš„å¹³å‡åˆ†"""
    total_score = 0
    for each in grades.values():
        total_score += int(each)
    average = f"(total_score/len(grades)):.1f"
    return average

# åŠŸèƒ½4ï¼šæ‰¾å‡ºæœ€é«˜åˆ†å’Œæœ€ä½åˆ†
def find_extremes():
    """è¿”å›æœ€é«˜åˆ†å’Œæœ€ä½åˆ†çš„å­¦ç”Ÿ"""
    highest = []
    lowest = []
    min_score = min(grades.values())
    max_score = max(grades.values())
    for every in grades:
        if grades[every] == min_score:
            lowest.append(every)
        elif grades[every] == max_score:
            highest.append(every)
    return highest, lowest

# åŠŸèƒ½5ï¼šæ˜¾ç¤ºæ‰€æœ‰å­¦ç”Ÿ
def show_all():
    """æ˜¾ç¤ºæ‰€æœ‰å­¦ç”Ÿæˆç»©"""
    print("\næ‰€æœ‰å­¦ç”Ÿæˆç»©:")
    for i in grades:
        print(f"{i}:{grades[i]}")

print("âœ… å­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿåˆå§‹åŒ–å®Œæˆï¼")
# é¡¹ç›®1ï¼šæµ‹è¯•ä»£ç 

print("=== å¼€å§‹æµ‹è¯• ===")

# æ·»åŠ ä¸€äº›æµ‹è¯•æ•°æ®
add_student("å¼ ä¸‰", 85)
add_student("æå››", 92)
add_student("ç‹äº”", 78)
add_student("èµµå…­", 88)

# æµ‹è¯•æŸ¥è¯¢
print(f"\næŸ¥è¯¢æµ‹è¯•:")
print(f"  å¼ ä¸‰çš„æˆç»©: {query_score('å¼ ä¸‰')}")
print(f"  æŸ¥è¯¢ä¸å­˜åœ¨çš„å­¦ç”Ÿ: {query_score('é’±ä¸ƒ')}")

# æ˜¾ç¤ºæ‰€æœ‰
show_all()

# è®¡ç®—å¹³å‡åˆ†
print(f"\nç»Ÿè®¡ä¿¡æ¯:")
print(f"  å¹³å‡åˆ†: {calculate_average():.1f}")

# æ‰¾å‡ºæœ€é«˜æœ€ä½åˆ†
highest, lowest = find_extremes()
print(f"  æœ€é«˜åˆ†: {highest}")
print(f"  æœ€ä½åˆ†: {lowest}")

print("\nâœ… æµ‹è¯•å®Œæˆï¼")
````
