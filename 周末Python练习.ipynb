````python
# ç»ƒä¹ 2.3ï¼šå­—å…¸éå†

student_scores = {
    "å¼ ä¸‰": 85,
    "æå››": 92,
    "ç‹äº”": 78,
    "èµµå…­": 95,
    "å­™ä¸ƒ": 88
}

print("å­¦ç”Ÿæˆç»©å­—å…¸:", student_scores)

# 1. éå†æ‰€æœ‰é”®ï¼ˆå­¦ç”Ÿå§“åï¼‰
print("\n1. æ‰€æœ‰å­¦ç”Ÿå§“å:")
for name in student_scores:
    print(f"   {name}")

# 2. éå†æ‰€æœ‰å€¼ï¼ˆæˆç»©ï¼‰
print("\n2. æ‰€æœ‰æˆç»©:")
for score in student_scores.values():
    print(f"   {score}")

# 3. éå†é”®å€¼å¯¹
print("\n3. å­¦ç”Ÿå§“åå’Œæˆç»©:")
for name, score in student_scores.items():
    print(f"   {name}: {score}")

# 4. è®¡ç®—å¹³å‡æˆç»©
total_score = 0
student_count = 0

for score in student_scores.values():
    total_score += score
    student_count += 1

average_score = total_score / student_count if student_count > 0 else 0
print(f"\n4. å¹³å‡æˆç»©: {average_score:.2f}")

# 5. æ‰¾å‡ºæœ€é«˜åˆ†å’Œæœ€ä½åˆ†
highest_score = 0
lowest_score = 100
top_student = ""
bottom_student = ""

for name, score in student_scores.items():
    if score > highest_score:
        highest_score = score
        top_student = name
    if score < lowest_score:
        lowest_score = score
        bottom_student = name

print(f"\n5. æˆç»©ç»Ÿè®¡:")
print(f"   æœ€é«˜åˆ†: {top_student} ({highest_score}åˆ†)")
print(f"   æœ€ä½åˆ†: {bottom_student} ({lowest_score}åˆ†)")

print("\n" + "="*50 + "\n")
````

````python
# ç»ƒä¹ 2.4ï¼šå­—å…¸ä¸åˆ—è¡¨ç»“åˆ

# åœºæ™¯ï¼šå¤šä¸ªå­¦ç”Ÿï¼Œæ¯ä¸ªå­¦ç”Ÿæœ‰å¤šé—¨è¯¾ç¨‹æˆç»©
# ä½¿ç”¨å­—å…¸åµŒå¥—åˆ—è¡¨çš„ç»“æ„
students_grades = {
    "å¼ ä¸‰": [85, 90, 88],  # è¯­æ–‡ã€æ•°å­¦ã€è‹±è¯­
    "æå››": [92, 85, 91],
    "ç‹äº”": [78, 88, 85]
}

print("å­¦ç”Ÿå„ç§‘æˆç»©:")
for name, grades in students_grades.items():
    print(f"  {name}: {grades}")

# 1. è®¡ç®—æ¯ä¸ªå­¦ç”Ÿçš„æ€»åˆ†
print("\n1. å­¦ç”Ÿæ€»åˆ†:")
for name, grades in students_grades.items():
    total = 0
    for grade in grades:
        total += grade
    print(f"  {name}: {total}åˆ†")

# 2. è®¡ç®—æ¯é—¨è¯¾çš„å¹³å‡åˆ†
# å…ˆç»Ÿè®¡æ¯é—¨è¯¾çš„åˆ†æ•°æ€»å’Œå’Œäººæ•°
course_totals = [0, 0, 0]  # ä¸‰é—¨è¯¾çš„æ€»åˆ†
student_count = len(students_grades)

for grades in students_grades.values():
    for i in range(len(grades)):
        course_totals[i] += grades[i]

print("\n2. å„ç§‘å¹³å‡åˆ†:")
course_names = ["è¯­æ–‡", "æ•°å­¦", "è‹±è¯­"]
for i in range(len(course_names)):
    average = course_totals[i] / student_count
    print(f"  {course_names[i]}: {average:.1f}åˆ†")

# 3. æ‰¾å‡ºæ¯é—¨è¯¾çš„æœ€é«˜åˆ†
print("\n3. å„ç§‘æœ€é«˜åˆ†:")
for course_index in range(3):  # ä¸‰é—¨è¯¾
    max_score = 0
    top_student = ""
    
    for name, grades in students_grades.items():
        if grades[course_index] > max_score:
            max_score = grades[course_index]
            top_student = name
    
    print(f"  {course_names[course_index]}: {top_student} ({max_score}åˆ†)")

print("\n" + "="*50 + "\n")
````

````python
# é¡¹ç›®1ï¼šç®€æ˜“ç”µè¯ç°¿
# è¦æ±‚ï¼šä½¿ç”¨å­—å…¸å­˜å‚¨å§“åå’Œç”µè¯ï¼Œå®ç°å¢åˆ æ”¹æŸ¥åŠŸèƒ½

# åˆå§‹åŒ–ç”µè¯ç°¿
phone_book = {
    "å¼ ä¸‰": "13800138000",
    "æå››": "13900139000"
}

print("=== ç®€æ˜“ç”µè¯ç°¿ç®¡ç†ç³»ç»Ÿ ===")

# ä¸»å¾ªç¯
while True:
    print("\nå½“å‰ç”µè¯ç°¿è”ç³»äºº:")
    if len(phone_book) == 0:
        print("  (ç©º)")
    else:
        for name, phone in phone_book.items():
            print(f"  {name}: {phone}")
    
    print("\nè¯·é€‰æ‹©æ“ä½œ:")
    print("  1. æ·»åŠ è”ç³»äºº")
    print("  2. æŸ¥æ‰¾è”ç³»äºº")
    print("  3. ä¿®æ”¹è”ç³»äººç”µè¯")
    print("  4. åˆ é™¤è”ç³»äºº")
    print("  5. é€€å‡ºç³»ç»Ÿ")
    
    choice = input("è¯·è¾“å…¥é€‰é¡¹(1-5): ")
    
    if choice == "1":
        # æ·»åŠ è”ç³»äºº
        name = input("è¯·è¾“å…¥è”ç³»äººå§“å: ")
        if name in phone_book:
            print(f"è”ç³»äºº'{name}'å·²å­˜åœ¨ï¼Œç”µè¯æ˜¯: {phone_book[name]}")
        else:
            phone = input("è¯·è¾“å…¥ç”µè¯å·ç : ")
            phone_book[name] = phone
            print(f"å·²æ·»åŠ è”ç³»äºº: {name}")
    
    elif choice == "2":
        # æŸ¥æ‰¾è”ç³»äºº
        name = input("è¯·è¾“å…¥è¦æŸ¥æ‰¾çš„è”ç³»äººå§“å: ")
        if name in phone_book:
            print(f"{name}: {phone_book[name]}")
        else:
            print(f"æœªæ‰¾åˆ°è”ç³»äºº: {name}")
    
    elif choice == "3":
        # ä¿®æ”¹è”ç³»äººç”µè¯
        name = input("è¯·è¾“å…¥è¦ä¿®æ”¹çš„è”ç³»äººå§“å: ")
        if name in phone_book:
            new_phone = input(f"è¯·è¾“å…¥'{name}'çš„æ–°ç”µè¯å·ç (åŸå·ç : {phone_book[name]}): ")
            phone_book[name] = new_phone
            print(f"å·²æ›´æ–°è”ç³»äºº: {name}")
        else:
            print(f"æœªæ‰¾åˆ°è”ç³»äºº: {name}")
    
    elif choice == "4":
        # åˆ é™¤è”ç³»äºº
        name = input("è¯·è¾“å…¥è¦åˆ é™¤çš„è”ç³»äººå§“å: ")
        if name in phone_book:
            confirm = input(f"ç¡®è®¤åˆ é™¤'{name}'å—?(Y/N): ")
            if confirm.lower() == "y":
                del phone_book[name]
                print(f"å·²åˆ é™¤è”ç³»äºº: {name}")
            else:
                print("å–æ¶ˆåˆ é™¤")
        else:
            print(f"æœªæ‰¾åˆ°è”ç³»äºº: {name}")
    
    elif choice == "5":
        # é€€å‡ºç³»ç»Ÿ
        print("æ„Ÿè°¢ä½¿ç”¨ç”µè¯ç°¿ï¼Œå†è§ï¼")
        break
    
    else:
        print("æ— æ•ˆé€‰é¡¹ï¼Œè¯·é‡æ–°è¾“å…¥ï¼")
````

````python
# é¡¹ç›®2ï¼šå•è¯é¢‘ç‡ç»Ÿè®¡
# è¦æ±‚ï¼šç»Ÿè®¡ä¸€æ®µæ–‡æœ¬ä¸­æ¯ä¸ªå•è¯å‡ºç°çš„æ¬¡æ•°

# ç¤ºä¾‹æ–‡æœ¬
text = "python is a powerful programming language python is easy to learn python is widely used"

print("åŸæ–‡:")
print(f"  '{text}'")

# 1. å°†æ–‡æœ¬åˆ†å‰²æˆå•è¯åˆ—è¡¨
words = text.split()
print(f"\nå•è¯åˆ—è¡¨: {words}")

# 2. ç»Ÿè®¡æ¯ä¸ªå•è¯å‡ºç°çš„æ¬¡æ•°
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("\nå•è¯å‡ºç°æ¬¡æ•°:")
for word, count in word_count.items():
    print(f"  {word}: {count}æ¬¡")

# 3. æ‰¾å‡ºå‡ºç°æœ€å¤šçš„å•è¯
max_count = 0
most_common_word = ""

for word, count in word_count.items():
    if count > max_count:
        max_count = count
        most_common_word = word

print(f"\nå‡ºç°æœ€å¤šçš„å•è¯: '{most_common_word}' ({max_count}æ¬¡)")

# 4. æ‰©å±•ï¼šæŒ‰å‡ºç°æ¬¡æ•°æ’åºæ˜¾ç¤º
print("\næŒ‰å‡ºç°æ¬¡æ•°æ’åº:")

# å°†å­—å…¸è½¬æ¢ä¸ºåˆ—è¡¨è¿›è¡Œæ’åº
word_items = []
for word, count in word_count.items():
    word_items.append([word, count])

# ä½¿ç”¨ç®€å•çš„æ’åºæ–¹æ³•ï¼ˆå†’æ³¡æ’åºæ€è·¯ï¼‰
for i in range(len(word_items)):
    for j in range(len(word_items) - 1):
        if word_items[j][1] < word_items[j+1][1]:  # æŒ‰å‡ºç°æ¬¡æ•°é™åºæ’åˆ—
            # äº¤æ¢ä½ç½®
            temp = word_items[j]
            word_items[j] = word_items[j+1]
            word_items[j+1] = temp

# æ˜¾ç¤ºæ’åºç»“æœ
for item in word_items:
    print(f"  {item[0]}: {item[1]}æ¬¡")

# 5. æ‰©å±•ï¼šå¤„ç†æ ‡ç‚¹ç¬¦å·ï¼ˆå¦‚æœæ–‡æœ¬åŒ…å«æ ‡ç‚¹ï¼‰
print("\n" + "="*50)
print("è¿›é˜¶æŒ‘æˆ˜ï¼šå¤„ç†å¸¦æ ‡ç‚¹çš„æ–‡æœ¬")

text_with_punctuation = "Python, is a powerful language. Python is easy! Python is widely used."
print(f"\nå¸¦æ ‡ç‚¹çš„æ–‡æœ¬: '{text_with_punctuation}'")

# å»é™¤æ ‡ç‚¹ç¬¦å·
import string

clean_words = []
current_word = ""

for char in text_with_punctuation:
    if char.isalpha():  # å¦‚æœæ˜¯å­—æ¯
        current_word += char.lower()  # è½¬ä¸ºå°å†™
    elif current_word:  # å¦‚æœä¸æ˜¯å­—æ¯ä¸”current_wordä¸ä¸ºç©º
        clean_words.append(current_word)
        current_word = ""

# å¤„ç†æœ€åä¸€ä¸ªå•è¯
if current_word:
    clean_words.append(current_word)

print(f"\næ¸…æ´—åçš„å•è¯: {clean_words}")
````
