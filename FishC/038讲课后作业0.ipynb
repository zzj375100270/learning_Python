网站注册简易版，保存用户名和密码，没有重复

```python
codes = {"小甲鱼":"I_love_FishC","不二如是":"FishC5201314"}
import hashlib
while True:
    name = input("请输入用户名：")
    if not name:
        print("请输入正确的用户名！")
        continue
    while True:
        if name in codes:
            print("该用户名已被注册！")
            name = input("请重新输入用户名：")
        else:
            break
    code = input("请输入密码：")
    codes[name] = code
    break
print("="*20)
print("目前已注册的用户有：")
for each in codes:
    result = hashlib.md5(bytes(codes[each],"utf-8"))
    print(each,'：',result.hexdigest())
```
