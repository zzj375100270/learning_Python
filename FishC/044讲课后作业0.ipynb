Fisher-Yates洗牌算法
对于该算法的实现描述如下（假设需要打乱 N 个数）：
1.写下从 1 到 N 的数字
2.获取一个 1 到剩下数字（包括这个数字）的随机数 k
3.从低位开始，取出第 k 个数字（这个数字还没有被取出），把它写在独立的一个列表的最后一位
4.重复前两个步骤，直到所有数据都被取出
5.现在，新的列表中就是打乱了顺序的结果
 请封装一个 fy_shuffle() 函数。
 参数1：可迭代对象类型，存储原始的数据
 参数2：整数类型，指定洗牌的次数
 返回值：列表类型，存储打乱后的数据
 要求：请按 Fisher-Yates 要求的算法步骤实现。

```python
import random
old_list = input("请输入需要打乱的序列：")
times = int(input("请输入需要打乱的次数："))

def fy_shuffle(old,times):
    time = 1
    while time <= times:
        old = list(old_list)
        s = len(old)
        new = []    
        while s > 0:
            i = random.randint(1,s)
            new.append(old.pop(i-1))
            s -= 1
        print(f"第{time}次打乱后的结果：{"".join(new)}")
        time += 1
    print(f"最终的结果是：{"".join(new)}")

fy_shuffle(old_list,times)
```
