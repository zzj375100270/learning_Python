在软件行业里，类似这种“项目经理要求这里运行缓慢，好让客户给钱优化”的梗，大家应该有听说过吧……
现在请大家编写一个装饰器，并将其放到适当的位置，目的是让代码 1 秒钟打印一个结果。

```python
import time

def time_plus(my_func):
    def plus_1s(*args, **kwargs):
        result = my_func(*args, **kwargs)
        time.sleep(1)
        return result
    return plus_1s


def fib():
    back1, back2 = 0, 1
    @time_plus 
    def func():
        nonlocal back1, back2
        back1, back2 = back2, back1 + back2
        print(back1, end=' ')
    return func


def get_fib(n):
    f = fib()
    for i in range(n):
        f()
    
n = int(input("请输入需要获取的斐波那契数："))
get_fib(n)
```
