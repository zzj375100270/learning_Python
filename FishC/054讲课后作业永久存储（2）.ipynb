```python
# 4.请编写代码，使用 pathlib 模块，列举出当前目录下的所有文件的尺寸。
from pathlib import Path
idx = Path.cwd()
print(f"当前路径：{idx}")
file_in = [x for x in Path(idx).iterdir() if x.is_file()]
for each in file_in:
    print(f"文件名：{Path(each).name}，文件尺寸：{Path(each).stat().st_size}")
```

```python
# 5.请编写代码，使用 pathlib 模块，列举出当前目录下的所有文件的修改时间。
from pathlib import Path
import time
idx = Path.cwd()
print(f"当前路径：{idx}")
file_in = [x for x in Path(idx).iterdir() if x.is_file()]
for each in file_in:
    print(f"文件名：{each.name}，修改时间：{time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(each.stat().st_mtime))}")
```

```python
# 6.请编写代码，使用 pathlib 模块，获取指定路径（文件夹）中所有的文件名，并按文件尺寸按从小到大进行排序。
from pathlib import Path
file_in = sorted([x for x in Path().iterdir() if x.is_file()],key=lambda x:x.stat().st_size)
for each in file_in:
    print(f"{each.name}({each.stat().st_size}字节)")
```

```python
# 8. 请编写代码，使用 pathlib 模块，找出指定路径（文件夹）中最后被修改的文件。
from pathlib import Path
import time
file_in = sorted([x for x in Path().iterdir() if x.is_file()],key=lambda x:x.stat().st_mtime)
print(f"'修改时间 -> {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(file_in[-1].stat().st_mtime))}','文件名 -> {file_in[-1].name}'")
```

```python
# 9.将8改为一句话模式
from pathlib import Path
import time
print(f"'修改时间 -> {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(sorted([x for x in Path().iterdir() if x.is_file()],key=lambda x:x.stat().st_mtime)[-1].stat().st_mtime))}','文件名 -> {sorted([x for x in Path().iterdir() if x.is_file()],key=lambda x:x.stat().st_mtime)[-1].name}'")
```
