##字典版鱼C影评小程序
需要存放的数据如下：
1.电影名称
2.上映时间
3.导演（可能多人）
4.主演（可能多人）
5.评分

```python
print('''欢迎进入鱼C影评小程序
1.数据录入
2.查询数据
3.退出程序''')
is_running = True
movies = {}
while is_running:
    func = input("请输入想要的功能（1/2/3）:")
    if func not in ['1','2','3']:
        print("输入错误！")
        continue
    elif func == '3':
        print("感谢您的使用！")
        break
    elif func == '1':
        while True:
            name = input("请输入电影名称：")
            date = input("请输入上映日期：")
            director = input("请输入导演名字（多人请用 / 分隔）：")
            actor = input("请输入演员名字（多人请用 / 分隔）：")
            score = input("请输入电影评分：")
            directors = [i.strip() for i in director.split('/')]
            actors = [j.strip() for j in actor.split('/')]
            movies[name] = {'名称':name,'日期':date,'导演':directors,'演员':actors,'评分':score}
            finish = input("请问是否继续录入（Y/N）:")
            if finish.lower() == 'y':
                continue
            elif finish.lower() == 'n':
                break
    elif func == '2':
        movie_name = input("请输入电影名字：")
        if movie_name in movies:
            movie = movies[movie_name]
            print(f"电影名称：{movie['名称']}\n上映日期：{movie['日期']}\n导演名单：{movie['导演']}\n演员名单：{movie['演员']}\n当前评分：{movie['评分']}")
            continue
        else:
            print("查无此片！")
            continue
```
