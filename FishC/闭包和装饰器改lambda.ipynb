#闭包函数和装饰器改写为lambda表达式

```python
def power(exp):
    def exp_of(base):
        return base ** exp
    return exp_of
```

```python
power = lambda exp:lambda base:base ** exp
square = power(2)
cube = power(3)
print(square(5))
print(cube(2))
print(cube(5))
```

```python
def add(func):
    def inner():
        x = func()
        return x + 1
    return inner
    
def cube(func):
    def inner():
        x = func()
        return x * x * x
    return inner
    
def square(func):
    def inner():
        x = func()
        return x * x
    return inner
    
@add
@cube
@square
def test():
    return 2
    
print(test())
```

```python
add = lambda func:lambda :func() + 1
cube = lambda func:lambda :func() ** 3
square = lambda func:lambda :func() ** 2
@add
@cube
@square
def test():
    return 2
print(test())
```
