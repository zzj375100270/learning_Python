小甲鱼frange()函数的设计精髓
1、使用stop=None作为哨兵，如果输入只有一位，那么在进行start和stop的分别赋值
2、pre这句代码的逻辑，利用字符串来判断小数位数

```python
def frange(start, stop=None, inc=0.1): #这句里的哨兵模式
    s = str(inc)
    pre = len(s[s.find('.')+1:])  #这句的利用字符串方法判断小数位数
    start += 0.0 
    if stop == None:
        stop = start + 0.0  
        start = 0.0
    while start < stop:
        yield round(start, pre)
        start += inc
```
