```python
def frange(start, stop=None, inc=0.1): #è¿™å¥é‡Œçš„å“¨å…µæ¨¡å¼
    s = str(inc)
    pre = len(s[s.find('.')+1:])  #è¿™å¥çš„åˆ©ç”¨å­—ç¬¦ä¸²æ–¹æ³•åˆ¤æ–­å°æ•°ä½æ•°
    start += 0.0 
    if stop == None:
        stop = start + 0.0  
        start = 0.0
    while start < stop:
        yield round(start, pre)
        start += inc
```
