```python
def print_cards():   #æ‰“å°æ‰€æœ‰ç‰Œï¼Œæ–¹ä¾¿ç”¨æˆ·è¾“å…¥æ—¶å¤åˆ¶ç²˜è´´
    cards = [["â™¥A","â™¥2","â™¥3","â™¥4","â™¥5","â™¥6","â™¥7","â™¥8","â™¥9","â™¥10","â™¥J","â™¥Q","â™¥K"],
            ["â™ A","â™ 2","â™ 3","â™ 4","â™ 5","â™ 6","â™ 7","â™ 8","â™ 9","â™ 10","â™ J","â™ Q","â™ K"],
            ["â™£A","â™£2","â™£3","â™£4","â™£5","â™£6","â™£7","â™£8","â™£9","â™£10","â™£J","â™£Q","â™£K"],
            ["â™¦A","â™¦2","â™¦3","â™¦4","â™¦5","â™¦6","â™¦7","â™¦8","â™¦9","â™¦10","â™¦J","â™¦Q","â™¦K"],
            ["â˜€","ğŸŒ™"]]
    for i in range(5):
        for each in cards[i]:
            print(each,end=" ")
        print()

def is_pair(cards):  #åˆ¤æ–­æ˜¯ä¸æ˜¯å¯¹å­
    if len(set(cards)) == 1:
        return True
    else:
        return False

def is_rocket(cards):    #åˆ¤æ–­æ˜¯ä¸æ˜¯ç«ç®­
    if set(cards) == {14,15}:
        return True
    else:
        return False
    
def is_triple(cards):    #åˆ¤æ–­æ˜¯ä¸æ˜¯ä¸‰è¿
    if len(set(cards)) == 1:
        return True
    else:
        return False
    
def is_bomb(cards):    #åˆ¤æ–­æ˜¯ä¸æ˜¯ç‚¸å¼¹
    if len(set(cards)) == 1:
        return True
    else:
        return False

def is_three_plus_one(cards):   #åˆ¤æ–­æ˜¯ä¸æ˜¯ä¸‰å¸¦ä¸€
    if len(set(cards)) == 2 and (cards.count(cards[0]) == 1 or cards.count(cards[0]) == 3):
        return True
    else:
        return False
    
def is_three_plus_two(cards):   #åˆ¤æ–­æ˜¯ä¸æ˜¯ä¸‰å¸¦äºŒ
    if len(cards) != 5:
        return False
    elif len(set(cards)) == 2 and (cards.count(cards[0]) == 2 or cards.count(cards[0]) == 3):
        return True
    else:
        return False
    
def is_four_plus_two(cards):   #åˆ¤æ–­æ˜¯ä¸æ˜¯å››å¸¦äºŒ
    if len(cards) != 6:
        return False
    if len(set(cards)) == 2 and (cards.count(cards[0]) == 2 or cards.count(cards[0]) == 4):
        return True
    else:
        return False
    
def is_continuous(cards):   #åˆ¤æ–­æ˜¯ä¸æ˜¯é¡ºå­
    if any(x in cards for x in (13,14,15)):
        return False
    elif len(cards) != len(set(cards)):
        return False
    elif len(cards) < 5:
        return False
    sorted_cards = sorted(cards)
    for i in range(1,len(sorted_cards)):
        if sorted_cards[i] - sorted_cards[i-1] != 1:
            return False
    return True

def is_double_continuous(cards):    #åˆ¤æ–­æ˜¯ä¸æ˜¯åŒé¡º
    if len(cards) < 6 or len(cards) % 2 != 0:
        return False
    if any(x in cards for x in (13,14,15)):
        return False
    sorted_cards = sorted(cards)
    if sorted_cards[0::2] != sorted_cards[1::2]:
        return False
    value = sorted_cards[0::2]
    for i in range(1,len(value)):
        if value[i] - value[i-1] != 1:
            return False
    return True

def is_triple_continuous(cards):    #åˆ¤æ–­æ˜¯ä¸æ˜¯ä¸‰é¡º
    if len(cards) < 6 or len(cards) % 3 != 0:
        return False
    if any(x in cards for x in (13,14,15)):
        return False
    sorted_cards = sorted(cards)
    if not sorted_cards[0::3] == sorted_cards[1::3] == sorted_cards[2::3]:
        return False
    value = sorted_cards[0::3]
    for i in range(1,len(value)):
        if value[i] - value[i-1] != 1:
            return False
    return True
    
def is_plane(cards):    #åˆ¤æ–­é£æœºå¸¦ç¿…è†€
    from collections import Counter
    if len(cards) < 8:
        return False
    count = Counter(cards)
    triples = [num for num,v in count.items() if v>=3 and num<=12]
    triples.sort()
    for start in range(len(triples)):
        for end in range(start+1,len(triples)):
            segment = triples[start:end+1]
            if not all(segment[i] + 1 == segment[i+1] for i in range(len(segment)-1)):
                continue
            k = len(segment)
            total = len(cards)
            if total == k * 4:
                return True
            if total == k * 5:
                remaining = {}
                for num,v in count.items():
                    if num in segment:
                        left = v - 3
                    else:
                        left = v
                    if left:
                        remaining[num] = left
                if sum(remaining.values()) == 2 * k and all(v == 2 for v in remaining.values()):
                    return True
    return False

def get_cards():    #æ¥æ”¶ç”¨æˆ·å‡ºçš„ç‰Œ
    info = input("è¯·å‡ºç‰Œ(ç©ºæ ¼é—´éš”,é€€å‡ºè¯·è¾“å…¥Q):")
    if info.upper() == "Q":
        return 0
    else:
        cards = info.split()
        return cards
    
def change_card(card):  #å°†ç”¨æˆ·çš„ç‰Œè½¬æ¢æˆæ•°å­—
    cards = []
    num = {"3":1,"4":2,"5":3,"6":4,"7":5,"8":6,"9":7,"10":8,"J":9,"Q":10,"K":11,"A":12,"2":13}
    for each in card:
        if len(each) != 1:
            cards.append(num.get(each[1:]))
        else:
            cards.append(14 if each == "ğŸŒ™" else 15)
    return cards

def check_type(cards):  #æ£€æŸ¥å‡ºçš„ç‰Œå±äºå“ªç§è§„åˆ™
    if len(cards) == 2:
        if is_pair(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šå¯¹ç‰Œ"
        elif is_rocket(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šç«ç®­"
        else:
            result = "ä¸ç¬¦åˆè§„åˆ™"
    elif len(cards) == 3:
        if is_triple(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šä¸‰å¼ ç‰Œç›¸åŒ"
        else:
            result = "ä¸ç¬¦åˆè§„åˆ™"
    elif len(cards) == 4:
        if is_bomb(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šç‚¸å¼¹"
        elif is_three_plus_one(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šä¸‰å¸¦ä¸€"
        else:
            result = "ä¸ç¬¦åˆè§„åˆ™"
    elif len(cards) >= 5:
        if is_three_plus_two(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šä¸‰å¸¦äºŒ"
        elif is_continuous(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šé¡ºå­"
        elif is_four_plus_two(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šå››å¸¦äºŒ"
        elif is_double_continuous(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šåŒé¡º"
        elif is_triple_continuous(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šä¸‰é¡º"
        elif is_plane(cards):
            result = "ç¬¦åˆè§„åˆ™ï¼šé£æœºå¸¦ç¿…è†€"
        else:
            result = "ä¸ç¬¦åˆè§„åˆ™"
    else:
        result = "ä¸ç¬¦åˆè§„åˆ™"
    return result

print_cards()
cards = get_cards()
while cards:
    cards = change_card(cards)
    result = check_type(cards)
    print(result)
    cards = get_cards()
```
