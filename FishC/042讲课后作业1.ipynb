##函数（2）
模拟栈的函数操作，push为压入，pop为取出，top为显示最上层，exit为退出

```python
stack = []
#定义函数 input_func()获取指令
def input_func():
    while True:
        func = input("请输入指令(push/pop/top/exit):")
        if func == "push":
            push()
        elif func == "pop":
            pop()
        elif func == "top":
            top()
        elif func == "exit":
            break
        else:
            print("指令错误！")

#定义函数 push()将数据压入栈
def push():
    pushed = input("请输入将要压入栈中的值：")
    stack.append(pushed)
    print("栈：")
    for i in range(len(stack)-1,-1,-1):
        print(stack[i])

#定义函数 pop()将数据取出栈
def pop():
    if stack:
        print(stack[-1])
        stack.pop()
    if stack:
        print("栈：")
        for i in range(len(stack)-1,-1,-1):
            print(stack[i])
    else:
        print("栈已空~")

#定义函数 top()显示栈顶数据
def top():
    if stack:
        print(stack[-1])
    else:
        print("栈已空~")

input_func()
```
