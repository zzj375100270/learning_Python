0. 爱因斯坦曾出过这样一道有趣的数学题：     
有一个长阶梯，若每步上 2 阶，最后剩 1 阶；若每步上 3 阶，最后剩 2 阶；若每步上 5 阶，最后剩 4 阶；若每步上 6 阶，最后剩 5 阶；只有每步上 7 阶，最后刚好一阶也不剩。     
求解该阶梯的最终数量？

```python
steps = 7
i = 1
FIND = False

while i < 100:
    if steps % 2 == 1 and steps % 3 == 2 and steps % 5 == 4 and steps % 6 == 5:
        FIND = True
        break
    else:
        steps = steps + 7
    i = i + 1

if FIND == True:
    print('阶梯数是：', steps)
else:
    print('在程序限定的范围内找不到答案！')
```
