0. 编写一个成绩评级程序，要求用户输入分数，程序返回对应的评级。
分数 < 60，D  
60 <= 分数 < 80，C  
80 <= 分数 < 90，B  
90 <= 分数 < 100，A  
分数 == 100，S

```python
score = int(input("请输入你的分数："))
if score < 0 or score > 100:
    print("别逗！")
elif score == 100:
    print("S")
elif score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 60:
    print("C")
else:
    print("D")
```

```python
while True:
    score = input("请输入你的分数：")
    if score == "e":
        break
    else:
        score = int(score)
        if score < 0 or score > 100:
            print("别逗！")
        elif score == 100:
            print("S")
        elif score >= 90:
            print("A")
        elif score >= 80:
            print("B")
        elif score >= 60:
            print("C")
        elif score >= 80:
            print("B")
        elif score >= 60:
            print("C")
        else:
            print("D")
```
