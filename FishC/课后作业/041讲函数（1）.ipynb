0. 请编写一个实现【注册】和【登陆】功能的代码，这次要求将不同的功能封装成独立的函数。       
程序要求如下：      
编写 4 个函数分别用于获取用户指令（get_int()）、注册（register()）、登陆（login()）、MD5加密（encrypt()）   
使用一个 Python 的字典作为数据库。      
注册时需验证用户名是否已存在于数据库        
登陆时需验证用户名和密码是否匹配        
密码保存需使用 MD5 加密     

```python
name_list = {}
import hashlib
#定义函数 get_int获取用户指令
def get_int():      
    while True:
        print("欢迎来到鱼C论坛~")
        print("="*25)
        print("1.注册")
        print("2.登录")
        print("3.退出")
        num = input("请输入指令：")
        if num == "1":
            register()
            continue
        elif num == "2":
            login()
            continue
        elif num == "3":
            print("="*25)
            break
        else:
            print("指令错误！")
#定义函数 register()注册
def register():
    print("="*25)
    while True:
        name = input("请输入用户名：")
        if name in name_list:
            print("该用户名已注册！")
            continue
        else:
            code = input("请输入密码：")
            codes = encrypt(code)
            name_list[name] = codes
            print("恭喜，注册成功~")
            break
            
#定义函数 login()登录
def login():
    print("="*25)
    name = input("请输入用户名：")
    while name not in name_list:
        print("该用户名不存在！")
        name = input("请重新输入用户名：")
    code = input("请输入密码：")
    while name_list[name] != encrypt(code):
        print("密码错误！")
        code = input("请重新输入密码：")
    print("恭喜，登录成功~")

#定义函数 encrypt()加密
def encrypt(code):
    return hashlib.md5(bytes(code,"utf-8")).hexdigest()

get_int()
```
