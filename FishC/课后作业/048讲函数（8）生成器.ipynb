利用生成器定义一个支持浮点数的 frange() 函数，其功能与 range() 函数相仿。       
默认的步进值为 0.1      
支持设置起始位置和结束位置      
当然也支持设置步进精度      
小甲鱼特别提示：由于 IEEE745 导致的浮点数误差问题，推荐大家考虑使用 round() 函数来修正其浮点误差的结果。

```python
def frange(*args):
    if len(args) > 3:
        result = "输入错误！"
        return result
    elif len(args) == 1:
        start = 0
        stop = args[0]
        step = 0.1
    elif len(args) == 2:
        start = args[0]
        stop = args[1]
        step = 0.1
    else:
        start = args[0]
        stop = args[1]
        step = args[2]
    num = start + 0.0
    while num < stop:
        yield round(num,1)
        num = round(num+step,1)


for i in frange(5,10,0.5):
    print(i)
```
