0. 请利用 __bases__ 属性，获取一个类的 MRO 顺序。       
这里我们获取 tkinter 模块中，Text 类的 MRO 顺序（这个 get_bases() 就是本题要求编写的函数，可以看到这里我们除了获取 MRO 顺序，还根据父类的层数，添加了相应数量的横杆）       
要求：      
不能使用 mro() 方法或 __mro__ 属性      
每深入一级父类，需要在前方添加一个横杆（-）

```python
def get_bases(text,level = 0):
    print(f"{'-' * level}{text}")
    for base in text.__bases__:
        if base != object:
            get_bases(base,level + 1)

import tkinter
get_bases(tkinter.Text)
```

最后编写一个搜索函数，通过提供文件名的方式搜索对应的文件位置和信息


```python
from unittest import result
from pathlib import Path
import time
base = r"D:\桌面\Python\learning_Python\FishC"
target = r"D:\桌面\Python\learning_Python\FishC\target"
file_list = [x for x in Path(target).rglob("*")]
target_list = [x.relative_to(base) for x in file_list]
for each in target_list:
    print(each)

class FileInfo:
    def __init__(self, file_path):
        self.path = Path(file_path)
        self.file_name = self.path.name
        self.size = self.path.stat().st_size
        self.ctime = self.path.stat().st_ctime
        self.mtime = self.path.stat().st_mtime
        self.atime = self.path.stat().st_atime
    
    def get_filename(self):
        return self.file_name
    
    def get_size(self):
        return self.size
    
    def get_path(self):
        return str(self.path)
    
    def get_creation_time(self):
        return time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(self.ctime))
    
    def get_modification_time(self):
        return time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(self.mtime))
    
    def get_access_time(self):
        return time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(self.atime))

def received_file():
    result = input("请输入要搜索的文件名：")
    return result

def search_file(keyword):
    result = [FileInfo(x) for x in file_list if keyword in x.relative_to(x)]
    return result

def print_result(search_result):
    if search_result:
        for times,files in enumerate(search_result):
            print(f"找到相关文件{times+1} -> {files.get_filename()}（{files.get_size()}字节）")
            print(f"位置：{files.get_path()}")
            print(f"创建时间：{files.get_creation_time()}")
            print(f"修改时间：{files.get_modification_time()}")
            print(f"访问时间：{files.get_access_time()}") 
        else:
            print("找不到相关文件！")
```
