0. 判断子序列       
给定字符串 s 和 t ，请编程判断 s 是否为 t 的子序列。        
字符串的子序列是原始字符串删除一些（也可以不删除）字符而不改变剩余字符相对位置形成的新字符串（例如，"ace" 是 "abcde" 的子序列，而 "aec" 则不是）。

```python
s = input("请输入字符串s：")        #双指针版本，这个好
t = input("请输入字符串t：")
i,j = 0,0
while i < len(s) and j < len(t):
    if s[i] == t[j]:
        i += 1
    j += 1
if i == len(s):
    print("字符串s是字符串t的子序列。")
else:
    print("字符串s不是字符串t的子序列。")
```

1. 给定一个字符串 s，请编程求出该字符串中的最大奇数。       
举例：      
输入：43383     
输出：43383         

输入：5926          
输出：59    

输入：966           
输出：9             

输入：64062     
输出：0

```python
old = input("输入：")
if not old.isdigit():
    print("输入错误！")
else:
    for i in range(len(old)-1,-1,-1):
        if int(old[i]) % 2 == 1:
            new = old[:i+1]
            break
        else:
            new = '0'
    print(f'输出：{new}')
```
