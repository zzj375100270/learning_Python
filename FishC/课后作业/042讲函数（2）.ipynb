0. 检测输入的中文字符串是否符合回文的语法，这个前面大家已经做过练习了，不过这一次我们将要放宽，只要该字的拼音一样，字不相同没有关系，比如 “前任只认钱” 可以符合要求。       
提示一：放宽要求后，只要文字的发音构成前后回文，即认定为符合要求。      
提示二：将汉字转换为对应拼音的方法 -> https://fishc.com.cn/thread-213439-1-1.html       
提示三：请将各个独立的功能封装为函数。

```python
from xpinyin import Pinyin
#定义函数 input_pinyin()将输入汉字转化为拼音
words = input("请输入一段话：")
while len(words) <= 1:
    words = input("字数太少，请重新输入：")
def input_pinyin(word):
    p = Pinyin()
    word_pinyin = p.get_pinyin(word).split("-")
    return word_pinyin
#定义函数 is_palindrome()判断是否是回文
def is_palindrome(words):
    left = 0
    right = len(words)-1
    while left < right:
        if words[left] != words[right]:
            return False
        else:
            left += 1
            right -= 1
    return True

if is_palindrome(input_pinyin(words)): 
    print(f"[{words}]是回文。")
else:
    print(f"[{words}]不是回文。")
```

1. 利用函数模拟创建【栈】的数据结构操作。       
知识解读：      
什么是栈呢？        
栈是一种具有 FILO 特性的数据结构，即先放入的数据反而后取出。        
比如有依次放入数据 250、251、255 到栈中，那么它们在栈中的存储结构如下：     
255     
251     
250     
此时，执行取出命令（pop），应该将最后压入栈的 255 优先取出，取出后栈中内容如下：        
251     
250     
同样道理，如果我们执行压栈命令（push），将数据 520 压入栈中，那么执行后，栈中内容如下：     
520     
251     
250     
题目要求：将压栈命令（push）、取出命令（pop）和打印栈的功能分别封装为独立的函数。       

```python
stack = []
#定义函数 input_func()获取指令
def input_func():
    while True:
        func = input("请输入指令(push/pop/top/exit):")
        if func == "push":
            push()
        elif func == "pop":
            pop()
        elif func == "top":
            top()
        elif func == "exit":
            break
        else:
            print("指令错误！")

#定义函数 push()将数据压入栈
def push():
    pushed = input("请输入将要压入栈中的值：")
    stack.append(pushed)
    print("栈：")
    for i in range(len(stack)-1,-1,-1):
        print(stack[i])

#定义函数 pop()将数据取出栈
def pop():
    if stack:
        print(stack[-1])
        stack.pop()
    if stack:
        print("栈：")
        for i in range(len(stack)-1,-1,-1):
            print(stack[i])
    else:
        print("栈已空~")

#定义函数 top()显示栈顶数据
def top():
    if stack:
        print(stack[-1])
    else:
        print("栈已空~")

input_func()
```
