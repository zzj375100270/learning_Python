0. 请改写附件的代码，使其功能变为计算 1000000 以内所有偶数的和.

```python
i = 0
x = 0
while i <= 1e6:
    if i % 2 == 0:
        x = x + i
        i = i + 2
print("1000000 以内所有偶数的和是" + str(x))
```

1. 舍罕王的失算     
请编程计算舍罕王应该给达依尔多少粒麦子？

```python
i = 0
s = 0
while i <= 63:
    wheats = 2 ** i
    s = s + wheats
    i = i + 1
print("舍罕王应该给达依尔", s, "粒麦子！")
```
