0. 请按要求编写一个企业员工管理程序。       
需要实现的要求：        
a. 员工分类：普通员工（Employee）、组长（Teamleader）、经理（Manager）      
b. 员工属性：姓名（name）、职位（job）、级别（grade）、工龄（year）、工号（uid）        
c. 员工方法：       
get_uid() -- 获取工号       
get_name() -- 获取姓名      
get_job() -- 获取职位       
get_grade() -- 获取级别     
get_year() -- 获取工龄      
salary() -- 统计并返回工资（计算方法如下）      
d. 员工工资计算方法：       
通员工：3000 + 500 * 级别 + 50 * 工龄       
组长：4000 + 800 * 级别 + 100 * 工龄        
经理：5000 + 1000 * (级别 + 工龄)       
e. 程序功能：       
录入员工数据功能（工号自动分配，由 10000 开始）     
查询功能，分为员工查询和职位查询（见程序实现截图）      
各类员工级别数量如下：      
普通员工：10 级     
组长：6 级      
经理：3 级      
升级功能：支持增加多级；当超过当前职位最高级别时，进行升职操作（无论增加多少个级别，只要发生升职，级别一律初始化为 1 级，见程序实现截图）."     
降级功能：支持减少多级；当超过当前职位最低级别时，进行降职操作（无论减少多少个级别，只要发生降职，级别一律初始化为当前职位最高级别，见程序实现截图）

```python
class Staff:
    def __init__(self, uid):
        self.uid = uid
        self.name = ""
        self.job = ""
        self.grade = 0
        self.year = 0
        self.salary = 0
        self.level = self.job + str(self.grade)
    def get_uid(self):
        return self.uid
    def get_name(self):
        return self.name
    def get_job(self):
        return self.job
    def get_grade(self):
        return self.grade
    def get_year(self):
        return self.year
    def get_salary(self):
        return self.salary
    def get_level(self):
        return self.level
    
class Employee(Staff):
    def __init__(self, uid):
        super().__init__(uid)
        self.job = "E"
        self.salary = 3000 + self.grade * 500 + self.year * 50

class Teamleader(Staff):
    def __init__(self, uid):
        super().__init__(uid)
        self.job = "T"
        self.salary = 4000 + self.grade * 800 + self.year * 100

class Manager(Staff):
    def __init__(self, uid):
        super().__init__(uid)
        self.job = "M"
        self.salary = 5000 + self.grade * 1000 + self.year * 1000

class Management:
    def __init__(self):
        self.staffs = []
        self.uid = 10000
        self.level = {"E1":1,"E2":2,"E3":3,"E4":4,"E5":5,"E6":6,"E7":7,"E8":8,"E9":9,"E10":10,
                      "T1":11,"T2":12,"T3":13,"T4":14,"T5":15,"T6":16,"M1":17,"M2":18,"M3":19}
        
    def add_staff(self):
        name = input("姓名：")
        job = input("职位（E.普通员工;T.组长;M.经理）：")
        if job == "E":
            staff = Employee(self.uid)
        elif job == "T":
            staff = Teamleader(self.uid)
        elif job == "M":
            staff = Manager(self.uid)
        else:
            print("职位输入错误")
            return
        staff.name = name
        year = int(input("工龄："))
        staff.year = year
        grade = int(input("级别："))
        staff.grade = grade
        self.staffs.append(staff)
        self.uid += 1
        print(f"录入成功！姓名：{staff.name}，工号：{staff.uid}，薪资：{staff.salary}")

    def staff_search(self):
        uid = int(input("请输入工号："))
        found = False
        for staff in self.staffs:
            if staff.get_uid() == uid:
                print(f"姓名：{staff.get_name()}")
                print(f"职位：{staff.get_job()}")
                print(f"级别：{staff.get_grade()}")
                print(f"工龄：{staff.get_year()}")
                print(f"薪资：{staff.get_salary()}")
                found = True
                break
        if not found:
            print("该工号不存在！")
            return
        
    def job_search(self):
        level_lst = {"E":"普通员工","T":"组长","M":"经理"}
        job = input("职位（E.普通员工;T.组长;M.经理）：")
        staff_lst = []
        staff_count = 0
        for staff in self.staffs:
            if staff.get_job() == job:
                staff_lst.append(staff)
                staff_count += 1
        if staff_count == 0:
            print(f"目前公司没有{level_lst[job]}！")
            return
        else:
            print(f"目前{level_lst[job]}共有{staff_count}人：")
            for staff in staff_lst:
                print(f"{staff.uid} - {staff.name}")

    def search(self):
        choice = input("1.员工查询;2.职位查询：")
        if choice == "1":
            self.staff_search()
        elif choice == "2":
            self.job_search()
        else:
            print("查询类型输入错误！")
            return
        
    def level_up(self):
        uid = int(input("请输入工号："))
        found = False
        for staff in self.staffs:
            if staff.get_uid() == uid:
                print(f"{staff.get_name()}，工号：{staff.get_uid()}，当前职位：{staff.get_level()}，当前薪资：{staff.get_salary()}") 
                found = True
                break
        if not found:
            print("该工号不存在！")
            return
```
