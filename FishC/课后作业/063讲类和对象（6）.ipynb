0. 请按要求编写一个企业员工管理程序。       
需要实现的要求：        
a. 员工分类：普通员工（Employee）、组长（Teamleader）、经理（Manager）      
b. 员工属性：姓名（name）、职位（job）、级别（grade）、工龄（year）、工号（uid）        
c. 员工方法：       
get_uid() -- 获取工号       
get_name() -- 获取姓名      
get_job() -- 获取职位       
get_grade() -- 获取级别     
get_year() -- 获取工龄      
salary() -- 统计并返回工资（计算方法如下）      
d. 员工工资计算方法：       
通员工：3000 + 500 * 级别 + 50 * 工龄       
组长：4000 + 800 * 级别 + 100 * 工龄        
经理：5000 + 1000 * (级别 + 工龄)       
e. 程序功能：       
录入员工数据功能（工号自动分配，由 10000 开始）     
查询功能，分为员工查询和职位查询（见程序实现截图）      
各类员工级别数量如下：      
普通员工：10 级     
组长：6 级      
经理：3 级      
升级功能：支持增加多级；当超过当前职位最高级别时，进行升职操作（无论增加多少个级别，只要发生升职，级别一律初始化为 1 级，见程序实现截图）."     
降级功能：支持减少多级；当超过当前职位最低级别时，进行降职操作（无论减少多少个级别，只要发生降职，级别一律初始化为当前职位最高级别，见程序实现截图）

```python
class Staff:
    def __init__(self, uid):
        self.uid = uid
        self.name = ""
        self.job = ""
        self.grade = 0
        self.year = 0
    def get_uid(self):
        return self.uid
    def get_name(self):
        return self.name
    def get_job(self):
        return self.job
    def get_grade(self):
        return self.grade
    def get_year(self):
        return self.year
    def get_salary(self):
        return self.salary()
    def get_level(self):
        return self.job + str(self.grade)
    def salary(self):
        pass

class Employee(Staff):
    def __init__(self, uid):
        super().__init__(uid)
        self.job = "E"
    def salary(self):
        return 3000 + self.grade * 500 + self.year * 50

class Teamleader(Staff):
    def __init__(self, uid):
        super().__init__(uid)
        self.job = "T"
    def salary(self):
        return 4000 + self.grade * 800 + self.year * 100

class Manager(Staff):
    def __init__(self, uid):
        super().__init__(uid)
        self.job = "M"
    def salary(self):
        return 5000 + 1000 * (self.grade + self.year)

class Management:
    def __init__(self):
        self.staffs = []
        self.uid = 10000
        
    def add_staff(self):
        name = input("姓名：")
        if not name:
            print("姓名不能为空！")
            return
        
        job = input("职位（E.普通员工;T.组长;M.经理）：")
        if job not in ["E", "T", "M"]:
            print("职位输入错误！")
            return
        
        if job == "E":
            staff = Employee(self.uid)
        elif job == "T":
            staff = Teamleader(self.uid)
        else:
            staff = Manager(self.uid)
        
        staff.name = name
        
        while True:
            try:
                year = int(input("工龄："))
                if year < 0:
                    print("工龄不能为负数！")
                    continue
                break
            except ValueError:
                print("请输入正确的数字！")
        staff.year = year
        
        job_grade_max = {"E": 10, "T": 6, "M": 3}
        while True:
            try:
                grade = int(input("级别："))
                if grade < 1 or grade > job_grade_max[job]:
                    print(f"级别输入错误！{job_grade_max[job]}级为最高级别")
                    continue
                break
            except ValueError:
                print("请输入正确的数字！")
        staff.grade = grade
        
        self.staffs.append(staff)
        self.uid += 1
        print(f"录入成功！姓名：{staff.name}，工号：{self.uid-1}，薪资：{staff.get_salary()}")

    def staff_search(self):
        while True:
            try:
                uid = int(input("请输入工号："))
                break
            except ValueError:
                print("请输入正确的数字！")
        
        found = False
        for staff in self.staffs:
            if staff.get_uid() == uid:
                print(f"姓名：{staff.get_name()}")
                print(f"职位：{staff.get_job()}")
                print(f"级别：{staff.get_grade()}")
                print(f"工龄：{staff.get_year()}")
                print(f"薪资：{staff.get_salary()}")
                found = True
                break
        if not found:
            print("该工号不存在！")
            return
        
    def job_search(self):
        level_lst = {"E":"普通员工","T":"组长","M":"经理"}
        job = input("职位（E.普通员工;T.组长;M.经理）：")
        if job not in level_lst:
            print("职位输入错误！")
            return
        
        staff_lst = []
        staff_count = 0
        for staff in self.staffs:
            if staff.get_job() == job:
                staff_lst.append(staff)
                staff_count += 1
        if staff_count == 0:
            print(f"目前公司没有{level_lst[job]}！")
            return
        else:
            print(f"目前{level_lst[job]}共有{staff_count}人：")
            for staff in staff_lst:
                print(f"{staff.uid} - {staff.name}")

    def search(self):
        choice = input("1.员工查询;2.职位查询：")
        if choice == "1":
            self.staff_search()
        elif choice == "2":
            self.job_search()
        else:
            print("查询类型输入错误！")
            return
        
    def level_up(self):
        while True:
            try:
                uid = int(input("请输入工号："))
                break
            except ValueError:
                print("请输入正确的数字！")
        
        found = False
        for staff in self.staffs:
            if staff.get_uid() == uid:
                print(f"{staff.get_name()}，工号：{staff.get_uid()}，当前职位：{staff.get_level()}，当前薪资：{staff.get_salary()}") 
                found = True
                break
        if not found:
            print("该工号不存在！")
            return
        
        while True:
            try:
                increase_level = int(input("请输入需要增加的级数："))
                if increase_level < 1:
                    print("增加级数必须大于0！")
                    continue
                break
            except ValueError:
                print("请输入正确的数字！")
        
        job_grade_max = {"E": 10, "T": 6, "M": 3}
        current_job = staff.get_job()
        current_grade = staff.get_grade()
        new_grade = current_grade + increase_level
        
        if new_grade <= job_grade_max[current_job]:
            staff.grade = new_grade
            print(f"升级成功！")
            print(f"{staff.get_name()}，工号：{staff.get_uid()}，升级后职位：{staff.get_level()}，升级后薪资：{staff.get_salary()}")
        else:
            if current_job == "M":
                print("已经是最高职位，无法继续升职！")
                return
            if current_job == "E":
                new_staff = Teamleader(staff.uid)
            else:
                new_staff = Manager(staff.uid)
            new_staff.name = staff.get_name()
            new_staff.year = staff.get_year()
            new_staff.grade = 1
            old_salary = staff.get_salary()
            new_salary = new_staff.get_salary()
            change = new_salary - old_salary
            idx = self.staffs.index(staff)
            self.staffs[idx] = new_staff
            print("升级成功！")
            print(f"{new_staff.name}，工号：{new_staff.uid}，升级后职位：{new_staff.get_level()}，升级后薪资：{new_salary}({change:+.0f})")

    def level_down(self):
        while True:
            try:
                uid = int(input("请输入工号："))
                break
            except ValueError:
                print("请输入正确的数字！")
        
        found = False
        for staff in self.staffs:
            if staff.get_uid() == uid:
                print(f"{staff.get_name()}，工号：{staff.get_uid()}，当前职位：{staff.get_level()}，当前薪资：{staff.get_salary()}") 
                found = True
                break
        if not found:
            print("该工号不存在！")
            return
        
        while True:
            try:
                demote_level = int(input("请输入需要减少的级数："))
                if demote_level < 1:
                    print("减少级数必须大于0！")
                    continue
                break
            except ValueError:
                print("请输入正确的数字！")
        
        job_grade_max = {"E": 10, "T": 6, "M": 3}
        current_job = staff.get_job()
        current_grade = staff.get_grade()
        new_grade = current_grade - demote_level
        
        if new_grade >= 1:
            staff.grade = new_grade
            print("降级成功！")
            print(f"{staff.get_name()}，工号：{staff.get_uid()}，降级后职位：{staff.get_level()}，降级后薪资：{staff.get_salary()}")
        else:
            if current_job == "E":
                print("已经是最低职位，无法继续降职！")
                return
            if current_job == "T":
                new_staff = Employee(staff.uid)
            else:
                new_staff = Teamleader(staff.uid)
            new_staff.name = staff.get_name()
            new_staff.year = staff.get_year()
            new_staff.grade = job_grade_max[new_staff.job]
            old_salary = staff.get_salary()
            new_salary = new_staff.get_salary()
            change = new_salary - old_salary
            idx = self.staffs.index(staff)
            self.staffs[idx] = new_staff
            print("降级成功！")
            print(f"{new_staff.name}，工号：{new_staff.uid}，升级后职位：{new_staff.get_level()}，升级后薪资：{new_salary}({change:+.0f})")

def main():
    manage = Management()
    while True:
        print("1.录入;2.查询;3.升级;4.降级;5.退出：")
        choice = input("请输入您的选择：")
        if choice == "1":
            manage.add_staff()
        elif choice == "2":
            manage.search()
        elif choice == "3":
            manage.level_up()
        elif choice == "4":
            manage.level_down()
        elif choice == "5":
            print("谢谢使用！")
            break
        else:
            print("选择输入错误！")
```
