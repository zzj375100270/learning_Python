0. 请帮开餐馆的小花开发一个点餐程序，要求如下。     
菜品种类如下：      
1.洋葱炒牛肉（1 份洋葱、1 份牛肉）        
2.洋葱炒羊肉（1 份洋葱、1 份羊肉）        
3.煎蛋（2 份蛋）      
4.番茄炒蛋（2 份番茄、3 份蛋）        
5.土豆萝卜炖羊肉（2 份土豆、1份萝卜、2 份羊肉）       
肉类价格：      
鸡蛋：1 元 1 份     
牛肉：25 元 1 份        
羊肉：30 元 1 份        
菜类价格：      
洋葱：2 元 1 份     
番茄：2 元 1 份     
土豆：3 元 1 份     
萝卜：3 元 1 份     
题目要求：      
Memu 类用于实例化菜单       
Memu 类需要有 order() 方法，用于点菜        
Memu 类需要有 pay() 方法，用于统计最终价格      
需要考虑同时点多个菜品的需求        
需要考虑一个菜品点多次的需求        

```python
from collections import Counter
food = {
    "鸡蛋": 1,
    "牛肉": 25,
    "羊肉": 30,
    "洋葱": 2,
    "番茄": 2,
    "土豆": 3,
    "萝卜": 3
}

class Menu:
    def __init__(self):
        self.orders = []
        self.goods = []
    
    def order(self):
        print("客官想要吃点什么？")
        while True:
            get_orders = input("1.洋葱炒牛肉；2.洋葱炒羊肉；3.煎蛋；4.番茄炒蛋；5. 土豆萝卜炖羊肉：")
            order_list = get_orders.split()
            valid = True
            temp = []
            for each in order_list:
                if each in ["1","2","3","4","5"]:
                    temp.append(each)
                else:
                    print("客官点的菜有问题，请重新点")
                    valid = False
                    break
            if valid:
                self.orders.extend(temp)
                break
    
    def order_to_goods(self):
        for order in self.orders:
            if order == "1":
                self.goods.extend(["洋葱","牛肉"])
            elif order == "2":
                self.goods.extend(["洋葱","羊肉"])
            elif order == "3":
                self.goods.extend(["鸡蛋","鸡蛋"])
            elif order == "4":
                self.goods.extend(["鸡蛋","鸡蛋","鸡蛋","番茄","番茄"])
            elif order == "5":
                self.goods.extend(["土豆","土豆","萝卜","羊肉","羊肉"])

    def pay(self):
        self.goods = []
        self.order_to_goods()
        order_count = Counter(self.goods)
        total = 0
        for k , v in order_count.items():
            print(f"{k} {food[k]} * {v}")
            total += food[k] * v
        print(f"感谢惠顾，您一共消费了{total}元，欢迎下次光临~")

menu = Menu()
menu.order()
menu.pay()
```
