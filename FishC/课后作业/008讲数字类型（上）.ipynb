0. 表达式 0.1 + 0.1 + 0.1 - 0.3 看似乎很傻，小学生都会，但 Python 可能搞不定，如下图所示，结果是一个非常奇葩的数值……请使用恰当的方式计算出正确的结果。

```python
import decimal
x = decimal.Decimal('0.1')
y = decimal.Decimal('0.3')
print(x + x + x - y)
```

1. 模拟抛硬币的实验

```python
import random
counts = int(input('请输入抛硬币的次数：'))
i = 0
print("开始抛硬币实验：")
while i < counts:
    num = random.randint(0,1)
    if num % 2 == 1:
        print('正面',end=' ')
    else:
        print('反面',end=' ')
    i = i + 1
```
