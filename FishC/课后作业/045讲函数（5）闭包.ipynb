0. 请设计一个函数，每次调用传入一个参数，并返回所有参数的平均值。

```python
def make_avg():
    total = 0
    times = 0
    def cal(nums):
        nonlocal total,times
        total += nums
        times += 1
        print(f"{total/times:.1f}")
    return cal
avg = make_avg()
avg(5)
avg(3)
avg(7)
avg(19)
```
