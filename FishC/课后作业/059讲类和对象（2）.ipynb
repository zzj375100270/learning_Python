0. 请根据下图所示继承关系创建 A~H 共计 8 个类，其中，父类 A 和 B 均拥有 x 属性（A.x = 250，B.x = 520），其余子类均为 pass 填充，然后判断 类 H 的实例化对象 h 的 x 属性值应该是多少？
图片略

```python
class A:
    x = 250
class B:
    x = 520
class C(B):
    pass
class D(A, B):
    pass
class E:
    pass
class F(D):
    pass
class G(D):
    pass
class H(E, F, G):
    pass

h = H()
assert h.x == 250
```
