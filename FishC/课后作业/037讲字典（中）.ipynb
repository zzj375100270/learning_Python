0. 请利用字典类型，改进上一节课后作业的电影查询小程序。     
需要存放的数据如下：        
电影名称        
上映时间        
导演（可能多人）        
主演（可能多人）        
电影评分        

```python
print('''欢迎进入鱼C影评小程序
1.数据录入
2.查询数据
3.退出程序''')
is_running = True
movies = {}
while is_running:
    func = input("请输入想要的功能（1/2/3）:")
    if func not in ['1','2','3']:
        print("输入错误！")
        continue
    elif func == '3':
        print("感谢您的使用！")
        break
    elif func == '1':
        while True:
            name = input("请输入电影名称：")
            date = input("请输入上映日期：")
            director = input("请输入导演名字（多人请用 / 分隔）：")
            actor = input("请输入演员名字（多人请用 / 分隔）：")
            score = input("请输入电影评分：")
            directors = [i.strip() for i in director.split('/')]
            actors = [j.strip() for j in actor.split('/')]
            movies[name] = {'名称':name,'日期':date,'导演':directors,'演员':actors,'评分':score}
            finish = input("请问是否继续录入（Y/N）:")
            if finish.lower() == 'y':
                continue
            elif finish.lower() == 'n':
                break
    elif func == '2':
        movie_name = input("请输入电影名字：")
        if movie_name in movies:
            movie = movies[movie_name]
            print(f"电影名称：{movie['名称']}\n上映日期：{movie['日期']}\n导演名单：{movie['导演']}\n演员名单：{movie['演员']}\n当前评分：{movie['评分']}")
            continue
        else:
            print("查无此片！")
            continue
```

1. 请编写一个电话簿程序。

```python
print("欢迎进入鱼C电话簿")
phone_number = {}
while True:
    func = input("请输入命令（I：录入 / C：查询 / D：删除 / P：打印 / E：退出）：")
    if func.lower() == 'i':
        print("-- 录入模式 --")
        inputting = True
        while inputting:
            name = input("请输入姓名：")
            if name in phone_number:
                print(f"该姓名已录入，手机号码是：{phone_number[name]['number']}")
                change = input("请问是否修改（Y/N）：")
                if change.lower() == 'y':
                    change_number = input("请输入新的手机号码：")
                    while (not change_number.isdigit()) or len(change_number) != 11:
                        change_number = input("输入不合法，请重新输入：")
                        continue
                    phone_number[name]['number'] = change_number
            else:
                number = input("请输入手机号码")
                while (not number.isdigit()) or len(number) != 11:
                    number = input("输入不合法，请重新输入：")
                phone_number[name] = {'name':name,'number':number}
            go_on = input("请问是否继续录入（Y/N）：")
            if go_on.lower() == 'y':
                continue
            elif go_on.lower() == 'n':
                break
            else:
                print("输入错误！返回录入模式！")
                continue
        continue
    elif func.lower() == 'c':
        print("-- 查询模式 --")
        while True:
            check_name = input("请输入姓名：")
            if check_name in phone_number:
                print(f"{check_name}：{phone_number[check_name]['number']}")
            else:
                print("查无此人！")
            go_on_check = input("是否继续查询（Y/N）")
            if go_on_check.lower() == 'y':
                continue
            elif go_on_check.lower() == 'n':
                break
            else:
                print("输入错误！返回查询模式！")
                continue
        continue
    elif func.lower() == 'd':
        print("-- 删除模式 --")
        while True:
            del_name = input("请输入姓名：")
            if del_name in phone_number:
                del phone_number[del_name]
            else:
                print("查无此人！")
            go_on_del = input("是否继续删除（Y/N）：")
            if go_on_del.lower() == 'y':
                continue
            elif go_on_del.lower() == 'n':
                break
            else:
                print("输入错误，返回删除模式！")
        continue
    elif func.lower() == 'p':
        print("-- 打印模式 --")
        if not phone_number:
            print("电话簿为空！")
        for each in phone_number:
            print(f"{each}：{phone_number[each]['number']}")
        continue
    elif func.lower() == 'e':
        print("感谢使用鱼C电话簿！")
        break
    else:
        print("无效指令，请重新输入！")
```
