#055讲课后作业永久存储（下）

```python
from pathlib import Path
p = Path.cwd()
x = p.iterdir()
def new_glob(x):
    list_py = []
    for each in x:
        if each.suffix == ".py":
            list_py.append(each)
        elif each.is_dir():
            list_py.extend(new_glob(each.iterdir()))
    return list_py
print(new_glob(x))
```
