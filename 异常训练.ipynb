综合练习：简单的用户输入验证
写一个程序，不断让用户输入两个数字和一个运算符（+ - * /），计算结果。要求：

捕获输入非数字的 ValueError

捕获除零错误

捕获不支持的运算符（抛出 ValueError）

如果一切正常，输出结果

输入 q 退出

```python
def div():
    run = 1
    while run:
        x = input("请输入第一个数：")
        y = input("请输入第二个数：")
        try:
            x_num = float(x)
            y_num = float(y)
            result = x_num / y_num
        except ValueError:
            print("数字输入错误~")
        except ZeroDivisionError:
            print("除数不能为0~")
        else:
            print("计算成功~")
            print(f"x / y = {result}")
            run = 0
        finally:
            print("计算结束~")
```
