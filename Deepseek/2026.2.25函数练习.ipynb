```python
def get_grade(score):
    """函数作用:根据接收到的成绩判断其评分
       score:接收到的成绩
       返回:评分;如果分数不在范围内，则返回None
    """
    level = {(90,100):"A",(80,89):"B",(70,79):"C",(60,69):"D",(0,59):"F"}
    for each in level.keys():
        if each[0] <= score <= each[1]:
            return level.get(each)
    return None

print(get_grade(95))
print(get_grade(105))
print(get_grade(85))
print(get_grade(75))
print(get_grade(65))
print(get_grade(55))
```

5. 组合练习：函数工厂
写一个函数 make_converter(unit_from, unit_to)，它返回一个函数，该函数可以将数值从 unit_from 单位转换为 unit_to 单位。支持 "c"（摄氏度）、"f"（华氏度）、"k"（开尔文）之间的转换。

```python
def C_to_F(temp):
    """摄氏度转华氏度
       temp:摄氏度
       返回:华氏度"""
    return temp * 9/5 + 32
def F_to_C(temp):
    """华氏度转摄氏度
       temp:华氏度
       返回:摄氏度"""
    return (temp - 32) * 5/9
def C_to_K(temp):
    """摄氏度转开尔文
       temp:摄氏度
       返回:开尔文"""
    return temp + 273.15
def K_to_C(temp):
    """开尔文转摄氏度
       temp:开尔文
       返回:摄氏度"""
    return temp - 273.15
def F_to_K(temp):
    """华氏度转开尔文
       temp:华氏度
       返回:开尔文"""
    return (temp - 32) * 5/9 +273.15
def K_to_F(temp):
    """开尔文转华氏度
       temp:开尔文
       返回:华氏度"""
    return (temp - 273.15) * 9/5 +32
def make_converter(unit_from,unit_to):
    """函数作用为温度转换函数工厂
       unit_from:转换前温度单位
       unit_to:目标温度单位
       返回:温度转换函数
    """
    trans = {("c","f"):C_to_F,("f","c"):F_to_C,("c","k"):C_to_K,
             ("k","c"):K_to_C,("f","k"):F_to_K,("k","f"):K_to_F}
    return trans.get((unit_from.lower(),unit_to.lower()))
c_to_f = make_converter("c","f")
f_to_k = make_converter("F","k")
print(c_to_f(0))
print(f_to_k(32))
```
