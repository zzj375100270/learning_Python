```python
#序列函数练习
words = ['hello','world','python','programming']
for num,word in enumerate(words):
    print(f"{num}-{word.upper()}")
```

```python
names = ['张三','李四','王五']
math_scores = [85,92,78]
english_scores = [88,90,85]
for name,mscore,escore in zip(names,math_scores,english_scores):
    print(f"{name}的总分是：{mscore+escore}。")
```

```python
temperatures_c = [0,20,37,100]
list(map(lambda x:int(x*9/5+32),temperatures_c))
```

```python
numbers = [12,45,78,23,56,89,34,67]
list(filter(lambda x:50<=x<=80,numbers))
```

```python
sentence = "Python is a powerful programming language"
in_word = sentence.split()
for word1,length in zip(in_word,map(len,in_word)):
    print(f"单词{word1}的长度是{length}。")
list(filter(lambda x:len(x)>3,in_word))
```
