简易通讯录管理系统

```python
contacts = []

def add_contact():
    """添加联系人（支持覆盖）"""
    while True:
        print("添加过程中输入q退出")
        name = input("请输入姓名：")
        if name.lower() == "q":
            break
        existing = None
        for c in contacts:        # 查找是否已存在同名联系人
            if c["姓名"] == name:
                existing = c
                break
        if existing:    # 覆盖：直接修改现有联系人信息
            choice = input("姓名已存在，是否覆盖(y/n)：")
            if choice.lower() != "y":
                continue
            phone = input("请输入手机号码：")
            if phone.lower() == "q":
                break
            if len(phone) != 11 or not phone.isdigit():
                print("手机号格式有误，必须为11位数字")
                continue
            email = input("请输入邮箱：")
            if email.lower() == "q":
                break
            existing["手机号"] = phone
            existing["邮箱"] = email
            print("更新成功")
        else:   # 新增联系人
            phone = input("请输入手机号码：")
            if phone.lower() == "q":
                break
            if len(phone) != 11 or not phone.isdigit():
                print("手机号格式有误，必须为11位数字")
                continue
            email = input("请输入邮箱：")
            if email.lower() == "q":
                break
            contacts.append({"姓名": name, "手机号": phone, "邮箱": email})
            print("添加成功")

def delete_contact():       #删除联系人
    while True:
        print("删除过程中输入q退出")
        name = input("请输入要删除的联系人姓名：")
        if name.lower() == "q":
            break
        for contack in contacts:
            if contack["姓名"] == name:
                contacts.remove(contack)
                print(f"联系人 {name} 删除成功")
                break
        else:
            print("联系人不存在")
            continue

def modify_contact():       #修改联系人
    while True:
        print("修改过程中输入q退出")
        name = input("请输入要修改的联系人姓名：")
        if name.lower() == "q":
            break
        for contack in contacts:
            if contack["姓名"] == name:
                print("当前联系人信息：")
                print(contack)
                phone = input("请输入新的手机号（不修改请直接回车）：")
                email = input("请输入新的邮箱（不修改请直接回车）：")
                if phone:
                    contack["手机号"] = phone
                if email:
                    contack["邮箱"] = email
                print(f"联系人 {name} 修改成功")
                break
        else:
            print("联系人不存在")
            continue

def search_contact():       #查找联系人
    while True:
        print("查询过程中输入q退出")
        name = input("请输入要查询的联系人姓名：")
        if name.lower() == "q":
            break
        for contack in contacts:
            if contack["姓名"] == name:
                email_display = contack["邮箱"] if contack["邮箱"] else "(空)"
                the_contact = [contack]
                width = calc_width(the_contact)
                print(f"{"姓名".ljust(width[0])}  {"手机号".ljust(width[1])}  {"邮箱".ljust(width[2])}")
                print("-" * (sum(width) + 4))
                print(f"{contack["姓名"].ljust(width[0])}  {contack["手机号"].ljust(width[1])}  {email_display.ljust(width[2])}")
                break
        else:
            print("联系人不存在")
            continue
        
def calc_width(contacts):       #计算打印行宽
    first = ["姓名","手机号","邮箱"]
    width = [len(each) for each in first]
    for contack in contacts:
        width[0] = max(width[0],len(contack["姓名"]))
        width[1] = max(width[1],len(contack["手机号"]))
        email_display = contack["邮箱"] if contack["邮箱"] else "(空)"
        width[2] = max(width[2],len(email_display))
    return width

def show_all_contact():     #显示所有联系人
    if not contacts:
        print("暂无联系人")
    else:
        width = calc_width(contacts)
        print(f"{"姓名".ljust(width[0])}  {"手机号".ljust(width[1])}  {"邮箱".ljust(width[2])}")
        print("-" * (sum(width) + 4))
        for contack in contacts:
            email_display = contack["邮箱"] if contack["邮箱"] else "(空)"
            print(f"{contack["姓名"].ljust(width[0])}  {contack["手机号"].ljust(width[1])}  {email_display.ljust(width[2])}")
        
def save_to_file():     #保存到文件
    with open("contacts.txt","w",encoding="utf-8") as f:
        f.write("\n".join([f"{contack["姓名"]},{contack["手机号"]},{contack["邮箱"] if contack["邮箱"] else "(空)"}" for contack in contacts]))

def load_from_file():   #从文件加载
    file_name = "contacts.txt"
    try:
        with open(file_name,"r",encoding="utf-8") as f:
            lines = f.readlines()
            for line in lines:
                line = line.strip()
                if line:
                    name,phone,email = line.split(",")
                    found = False
                    for c in contacts:
                        if c["姓名"] == name:
                            c["手机号"] = phone
                            c["邮箱"] = email
                            found = True
                            break
                    if not found:
                        contact = {"姓名":name,"手机号":phone,"邮箱":email}
                        contacts.append(contact)
    except FileNotFoundError:
        print("文件不存在")
        return
    except Exception as e:
        print(f"读取文件时出错：{e}")
        return

def main():     #主菜单
    while True:
        print("1. 添加联系人")
        print("2. 删除联系人")
        print("3. 修改联系人")
        print("4. 查询联系人")
        print("5. 显示所有联系人")
        print("6. 保存到文件")
        print("7. 从文件加载")
        print("0. 退出")
        choice = input("请输入您的选择：")
        if choice == "1":
            add_contact()
        elif choice == "2":
            delete_contact()
        elif choice == "3":
            modify_contact()
        elif choice == "4":
            search_contact()
        elif choice == "5":
            show_all_contact()
        elif choice == "6":
            save_to_file()
        elif choice == "7":
            load_from_file()
        elif choice == "0":
            break
        else:
            print("无效选择，请重新输入")
```
