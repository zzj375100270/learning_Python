题目1：成绩乘法器
写一个闭包 make_multiplier(n)，它返回一个函数。这个返回的函数接收一个数字，返回这个数字乘以 n 的结果。

```python
def make_multiplier(n):
    def get_num(num):
        result = num*n
        return result
    return get_num

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))
print(triple(5))
```

题目2：对话记录器
写一个闭包 chat_memory()，它返回一个函数。这个返回的函数接收一句话，每次调用时打印“历史：”加上之前所有说过的话（包括当前这句），用“｜”连接。

```python
def chat_memory():
    memory = []
    def chatting(words):
        nonlocal memory
        memory.append(words)
        result = "|".join(memory)
        print(f"历史：{result}")
    return chatting

chat = chat_memory()
chat("你好")
chat("今天冷")
chat("注意保暖")
```

题目3：函数备注器
写一个装饰器 @add_note，它接收一个字符串参数。被装饰的函数执行时，先打印这个备注，再执行原函数。

```python
def add_note(words):
    def note(func):
        def notes(*args):
            result = func(*args)
            print(f"备注：{words}")
            return result
        return notes
    return note

@add_note("这是计算函数")
def add(a,b):
    return a+b

print(add(3,5))
```

题目4：防呆装饰器
写一个装饰器 @ensure_positive，它检查被装饰函数的第一个参数。如果第一个参数 ≤ 0，打印“参数必须为正数！”并返回 None；否则正常执行函数。

```python
def ensure_positive(func):
    def check_it(*args):
        if args[0] <= 0:
            print("参数必须为正数！")
        else:
            return func(*args)
    return check_it

@ensure_positive
def divide_10_by(x):
    return 10/x

print(divide_10_by(5))
print(divide_10_by(0))
print(divide_10_by(-2))
```

题目5：快速排序键
有一份商品数据，用 lambda 表达式完成以下排序需求：
请分别写出：
按价格从低到高排序（用 lambda）
按库存从高到低排序（用 lambda）
按“价格 × 库存”的总价值从高到低排序（用 lambda）

```python
products = [
    {"name": "笔记本", "price": 15, "stock": 100},
    {"name": "钢笔", "price": 8, "stock": 200},
    {"name": "橡皮", "price": 3, "stock": 500},
    {"name": "尺子", "price": 6, "stock": 150}
]

price_up = sorted(products,key=lambda product:product["price"])
print(price_up)

stock_down = sorted(products,key=lambda product:product["stock"],reverse=True)
print(stock_down)

value_down = sorted(products,key=lambda product:product["price"]*product["stock"],reverse=True)
print(value_down)
```

```python
def counter1():
    count = 0
    def add():
        nonlocal count
        count += 1
        return count
    return add
counter = counter1()
```
