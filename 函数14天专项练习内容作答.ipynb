ğŸ“… ç¬¬5å¤©ï¼šé€’å½’å…¥é—¨
ç›®æ ‡ï¼šç†è§£é€’å½’æ€æƒ³ï¼Œèƒ½å†™ç®€å•é€’å½’å‡½æ•°ã€‚

ä»»åŠ¡æ¸…å•ï¼š

é€’å½’æ±‚å’Œï¼šå†™ recursive_sum(n) è®¡ç®— 1+2+...+nã€‚

é€’å½’æ±‚é˜¶ä¹˜ï¼šfactorial(n)ã€‚

é€’å½’éå†åˆ—è¡¨ï¼šå†™ recursive_max(lst) è¿”å›åˆ—è¡¨ä¸­çš„æœ€å¤§å€¼ï¼ˆç”¨é€’å½’ï¼Œä¸èƒ½ç”¨å†…ç½®maxï¼‰ã€‚æç¤ºï¼šreturn max(lst[0], recursive_max(lst[1:])) ä½†è¦æ³¨æ„ç©ºåˆ—è¡¨æƒ…å†µã€‚

æäº¤ï¼šä»£ç å¹¶ç®€è¦è¯´æ˜é€’å½’åŸºå’Œé€’å½’æ­¥ã€‚

```python
def recrusive_sum(n):   #é€’å½’æ±‚å’Œ
    if n == 1:
        return 1
    return n + recrusive_sum(n-1)
print(recrusive_sum(10))

def factorial(n):   #é€’å½’æ±‚é˜¶ä¹˜
    if n == 1 or n == 0:
        return 1
    return n * factorial(n-1)
print(factorial(2))
print(factorial(5))
print(factorial(10))

def recursive_max(lst):     #é€’å½’æ±‚æœ€å¤§å€¼
    if not lst:
        return None
    if len(lst) == 1:
        return lst[0]
    return max(lst[0],recursive_max(lst[1:]))
print(recursive_max([]))
print(recursive_max([3]))
print(recursive_max([1,2,3,4,5]))
print(recursive_max([1,2,3,-1,-5,-4,5]))
```

æ‰©å±•ï¼šç»Ÿè®¡è¿™äº› .py æ–‡ä»¶çš„æ€»è¡Œæ•°ï¼ˆæ’é™¤ç©ºè¡Œï¼‰ï¼Œå‡½æ•° count_lines_in_py(directory)ã€‚

```python
direc = r"D:\æ¡Œé¢\Pathon\ç»ƒä¹ \FishC"
from pathlib import Path
def find_py_files(directory):
    files = []
    path = Path(directory)
    for each in path.iterdir():
        if each.suffix == ".py":
            files.append(each)
        if each.is_dir():
            files.extend(find_py_files(each))
    return files

def count_lines_in_py(directory):
    count = 0
    py_files = find_py_files(directory)
    for each in py_files:
        with open(each,"r",encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if line:
                    count += 1
    return count

print(find_py_files(direc))
print(f"æ€»å…±æœ‰{len(find_py_files(direc))}ä¸ªæ–‡ä»¶")
print(count_lines_in_py(direc))
```

å†™æ–æ³¢é‚£å¥‘å‡½æ•° fib(n)ï¼ˆé€’å½’ç‰ˆæœ¬ï¼‰ï¼Œç„¶åæ”¹å†™ä¸ºå¸¦è®°å¿†åŒ–çš„ç‰ˆæœ¬ fib_memo(n, memo={})ï¼Œæ¯”è¾ƒ n=35 æ—¶çš„è¿è¡Œé€Ÿåº¦ï¼ˆå¯ç”¨ time æ¨¡å—ç²—ç•¥æµ‹é‡ï¼‰ã€‚

æäº¤ï¼šä¸¤ç§æ–æ³¢é‚£å¥‘ä»£ç å’Œé€Ÿåº¦å¯¹æ¯”ã€‚

```python
def hanoi(n,A,B,C):     #æ±‰è¯ºå¡”
    if n == 1:
        print(f"{A} --> {C}")
    elif n > 1:
        hanoi(n-1,A,C,B)
        print(f"{A} --> {C}")
        hanoi(n-1,B,A,C)
    else:
        return None
hanoi(3,"x","y","z")

import time
def fib(n):     #æ–æ³¢é‚£å¥‘é€’å½’ç‰ˆ
    if n == 1 or n == 2:
        return 1
    return fib(n-1) + fib(n-2)

def fib_memo(n,memo={}):    #æ–æ³¢é‚£å¥‘é€’å½’è®°å¿†ç‰ˆ
    if n == 1 or n == 2:
        return 1
    if n not in memo:
        memo[n] = fib_memo(n-1) + fib_memo(n-2)
    return memo[n]

start = time.time()
fib(40)
end = time.time()
print(f"è€—æ—¶{end-start:.2f}ç§’")

start = time.time()
fib_memo(40)
end = time.time()
print(f"è€—æ—¶{end-start:.2f}ç§’")

def fib_memo1(n,memo=None):     #æ–æ³¢é‚£å¥‘é€’å½’è®°å¿†å“¨å…µç‰ˆï¼Œè¿™ç§æœ€å¥½
    if memo == None:
        memo = {}
    if n == 1 or n == 2:
        return 1
    if n not in memo:
        memo[n] = fib_memo1(n-1,memo) + fib_memo1(n-2,memo)
    return memo[n]
```
