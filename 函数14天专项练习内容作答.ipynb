```python
def is_pair(cards):     #åˆ¤æ–­å¯¹å­
    return len(cards) == 2 and len(set(cards)) == 1

def is_bomb(cards):     #åˆ¤æ–­ç‚¸å¼¹
    return len(cards) == 4 and len(set(cards)) == 1

def is_rocket(cards):     #åˆ¤æ–­ç«ç®­
    return len(cards) == 2 and set(cards) == {14,15}

def calc(operation:str,a:int|float,b:int|float):    #ç®€æ˜“è®¡ç®—å™¨
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        if b == 0:
            return None
        else:
            return a / b
    else:
        return None
```

```python
def add(a,b):
    """è¿”å›a+bçš„ç»“æœ"""
    return a + b
def sub(a,b):
    """è¿”å›a-bçš„ç»“æœ"""
    return a - b
def mul(a,b):
    """è¿”å›a*bçš„ç»“æœ"""
    return a * b
def div(a,b):
    """è¿”å›a/bçš„ç»“æœï¼Œå¦‚æœb=0ï¼Œè¿”å›None"""
    return a / b if b != 0 else None
def my_pow(a,b):
    """è¿”å›a**bçš„ç»“æœ"""
    return a ** b
def floor_div(a,b):
    """è¿”å›a//bçš„ç»“æœï¼Œå¦‚æœb=0ï¼Œè¿”å›None"""
    return a // b if b != 0 else None
def qu_yu(a,b):
    """è¿”å›a%bçš„ç»“æœï¼Œå¦‚æœb=0ï¼Œè¿”å›None"""
    return a % b if b != 0 else None

def calc(operation:str,a:int|float,b:int|float):
    """å­—å…¸æ˜ å°„ç‰ˆç®€æ˜“è®¡ç®—å™¨
       operationä¸ºè¿ç®—ç¬¦ï¼Œå¯ä»¥æ˜¯"+","-","*","/","**","//","%",å¦‚æœä¸æ˜¯è¿™äº›ç¬¦å·åˆ™è¿”å›None
       aå’Œbä¸ºæ•°å­—ï¼Œå¯ä»¥ä¸ºæ•´å‹å’Œæµ®ç‚¹æ•°"""
    operations = {"+":add,"-":sub,"*":mul,"/":div,"**":my_pow,"//":floor_div,"%":qu_yu}
    func = operations.get(operation)
    if func:
        return func(a,b)
    else:
        return None

print(calc.__doc__)
```
