```python
def is_pair(cards):     #åˆ¤æ–­å¯¹å­
    return len(cards) == 2 and len(set(cards)) == 1

def is_bomb(cards):     #åˆ¤æ–­ç‚¸å¼¹
    return len(cards) == 4 and len(set(cards)) == 1

def is_rocket(cards):     #åˆ¤æ–­ç«ç®­
    return len(cards) == 2 and set(cards) == {14,15}

def calc(operation:str,a:int|float,b:int|float):    #ç®€æ˜“è®¡ç®—å™¨
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        if b == 0:
            return None
        else:
            return a / b
    else:
        return None
```

```python
def counter(start=1,step=1):
    count = start
    def inner():
        nonlocal count
        result = count
        count += step
        return result
    return inner
func1 = counter()
func2 = counter(5,2)
print(func1())
print(func1())
print(func2())
print(func2())
```
