```python
def is_pair(cards):     #判断对子
    return len(cards) == 2 and len(set(cards)) == 1

def is_bomb(cards):     #判断炸弹
    return len(cards) == 4 and len(set(cards)) == 1

def is_rocket(cards):     #判断火箭
    return len(cards) == 2 and set(cards) == {14,15}

def calc(operation:str,a:int|float,b:int|float):    #简易计算器
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        if b == 0:
            return None
        else:
            return a / b
    else:
        return None
```

📅 第5天：递归入门
目标：理解递归思想，能写简单递归函数。

任务清单：

递归求和：写 recursive_sum(n) 计算 1+2+...+n。

递归求阶乘：factorial(n)。

递归遍历列表：写 recursive_max(lst) 返回列表中的最大值（用递归，不能用内置max）。提示：return max(lst[0], recursive_max(lst[1:])) 但要注意空列表情况。

提交：代码并简要说明递归基和递归步。

```python
def recrusive_sum(n):   #递归求和
    if n == 1:
        return 1
    return n + recrusive_sum(n-1)
print(recrusive_sum(10))

def factorial(n):   #递归求阶乘
    if n == 1 or n == 0:
        return 1
    return n * factorial(n-1)
print(factorial(2))
print(factorial(5))
print(factorial(10))

def recursive_max(lst):     #递归求最大值
    if not lst:
        return None
    if len(lst) == 1:
        return lst[0]
    return max(lst[0],recursive_max(lst[1:]))
print(recursive_max([]))
print(recursive_max([3]))
print(recursive_max([1,2,3,4,5]))
print(recursive_max([1,2,3,-1,-5,-4,5]))
```

任务清单：

用 pathlib 和递归，编写函数 find_py_files(directory)，返回该目录及其所有子目录下的 .py 文件路径列表（Path对象）。不能用 rglob，必须自己写递归。

扩展：统计这些 .py 文件的总行数（排除空行），函数 count_lines_in_py(directory)。

提交：代码及运行结果（可以指定一个小目录测试）。

```python
direc = r"D:\桌面\Pathon\练习\FishC"
from pathlib import Path
def find_py_files(directory):
    files = []
    path = Path(directory)
    for each in path.iterdir():
        if each.suffix == ".py":
            files.append(each)
        if each.is_dir():
            files.extend(find_py_files(each))
    return files

def count_lines_in_py(directory):
    count = 0
    py_files = find_py_files(directory)
    for each in py_files:
        with open(each,"r",encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if line:
                    count += 1
    return count

print(find_py_files(direc))
print(f"总共有{len(find_py_files(direc))}个文件")
print(count_lines_in_py(direc))
```

📅 第7天：递归挑战——汉诺塔与斐波那契
目标：加深递归理解，接触优化思想。

任务清单：

写汉诺塔函数 hanoi(n, A, B, C)，打印每一步移动。测试 n=3。

写斐波那契函数 fib(n)（递归版本），然后改写为带记忆化的版本 fib_memo(n, memo={})，比较 n=35 时的运行速度（可用 time 模块粗略测量）。

提交：两种斐波那契代码和速度对比。

```python
def hanoi(n,A,B,C):     #汉诺塔
    if n == 1:
        print(f"{A} --> {C}")
    elif n > 1:
        hanoi(n-1,A,C,B)
        print(f"{A} --> {C}")
        hanoi(n-1,B,A,C)
    else:
        return None
hanoi(3,"x","y","z")

import time
def fib(n):     #斐波那契递归版
    if n == 1 or n == 2:
        return 1
    return fib(n-1) + fib(n-2)

def fib_memo(n,memo={}):    #斐波那契递归记忆版
    if n == 1 or n == 2:
        return 1
    if n not in memo:
        memo[n] = fib_memo(n-1) + fib_memo(n-2)
    return memo[n]

start = time.time()
fib(40)
end = time.time()
print(f"耗时{end-start:.2f}秒")

start = time.time()
fib_memo(40)
end = time.time()
print(f"耗时{end-start:.2f}秒")

def fib_memo1(n,memo=None):     #斐波那契递归记忆哨兵版，这种最好
    if memo == None:
        memo = {}
    if n == 1 or n == 2:
        return 1
    if n not in memo:
        memo[n] = fib_memo1(n-1,memo) + fib_memo1(n-2,memo)
    return memo[n]
```

```python
def make_multiplier(n):
    def inner(x):
        return x * n
    return inner
mul_2 = make_multiplier(2)
print(mul_2(5))
mul_5 = make_multiplier(5)
print(mul_5(10))
```

```python
import time
from functools import wraps
def timer(func):
    @wraps(func)
    def inner(*args,**kwargs):
        start = time.time()
        result = func(*args,**kwargs)
        end = time.time()
        print(f"耗时{end-start:.5f}秒")
        return result
    return inner

@timer
def fib(n):
    if n == 1 or n == 2:
        return 1
    a = b = 1
    while n > 2:
        a,b = b,a+b
        n -= 1
    return b
fib(30)
```

```python
from functools import wraps
def repeat(n):
    def outer(func):
        @wraps(func)
        def inner(*args,**kwargs):
            return [func(*args,**kwargs) for _ in range(n)]
        return inner
    return outer

@repeat(3)
def fib(n):
    if n == 1 or n == 2:
        return 1
    a = b = 1
    while n > 2:
        a,b = b,a+b
        n -= 1
    return b
fib(30)
```

```python
from functools import wraps
def repeat(n):
    def outer(func):
        @wraps(func)
        def inner(*args,**kwargs):
            result = []
            for _ in range(n):
                result.append(func(*args,**kwargs))
            return result
        return inner
    return outer
```

📅 第10天：装饰器进阶
目标：掌握带参数的装饰器，理解 functools.wraps。

任务清单：

写一个带参数的装饰器 retry(max_attempts)，当被装饰函数抛出异常时，自动重试最多 max_attempts 次。如果都失败，则抛出最后一次异常。

将 timer 装饰器改进，使其能通过 @timer('ms') 指定时间单位（秒或毫秒）。

使用 functools.wraps 保留原函数的元信息（如 __name__、__doc__）。

提交：代码并演示。

```python
import time
from functools import wraps
def timer(output="s"):
    def outer(func):
        @wraps(func)
        def inner(*args,**kwargs):
            start = time.time()
            result = func(*args,**kwargs)
            end = time.time()
            if output.lower() == "ms":
                print(f"耗时{(end-start)*1000:.2f}毫秒")
            else:
                print(f"耗时{end-start:.5f}秒")
            return result
        return inner
    return outer
```
